2.1. Se estimó un modelo de regresión lineal donde el salario del CEO (salary) depende de las variables sales, roe y ros.

El modelo es:

salary = β0 + β1·sales + β2·roe + β3·ros + error

Cada coeficiente mide el efecto de la variable correspondiente sobre el salario, manteniendo constantes las demás.

La significancia individual de los regresores se evalúa mediante sus p-values:

Si p < 0,05 → variable significativa
Si p > 0,05 → variable no significativa

La significancia global del modelo se evalúa mediante el estadístico F:

Si el p-value del F es menor a 0,05, el modelo es globalmente significativo.

Pregunta 2.5- Usando la regresión del salario en función de sales, roe y ros, se distinguen los
grupos mediante las dummies; indus, finance, consprod y utility, que corresponden al tipo de
empresa. Realice mediante test de Chow, con 5% de significancia, si los modelos para cada
grupo son iguales.

```r
ceosal1_indus <- subset(ceosal1, indus == 1)
ceosal1_finance <- subset(ceosal1, finance == 1)
ceosal1_consprod <- subset(ceosal1, consprod == 1)
ceosal1_utility <- subset(ceosal1, utility == 1)

modelo <- lm(salary ~ sales + roe + ros, data = ceosal1)

modelo_indus <- lm(salary ~ sales + roe + ros, data = ceosal1_indus)
modelo_finance <- lm(salary ~ sales + roe + ros, data = ceosal1_finance)
modelo_consprod <- lm(salary ~ sales + roe + ros, data = ceosal1_consprod)
modelo_utility <- lm(salary ~ sales + roe + ros, data = ceosal1_utility)

SSR_r <- sum(residuals(modelo)^2)

SSR_indus <- sum(residuals(modelo_indus)^2)
SSR_finance <- sum(residuals(modelo_finance)^2)
SSR_consprod <- sum(residuals(modelo_consprod)^2)
SSR_utility <- sum(residuals(modelo_utility)^2)

SSR_ur <- SSR_indus + SSR_finance + SSR_consprod + SSR_utility

G <- 4
k <- 3
n <- nrow(ceosal1)

gl1 <- (G - 1) * (k + 1)
gl2 <- n - G * (k + 1)

F_est <- ((SSR_r - SSR_ur) / gl1) / (SSR_ur / gl2)

F_critico <- qf(0.95, gl1, gl2)

F_est
F_critico

if (F_est > F_critico) {
  print("Se rechaza H0")
} else {
  print("No se rechaza H0")
}
```

2.5. Se aplica un test de Chow para evaluar si el modelo de regresión es igual entre los distintos sectores económicos.

H0: los modelos son iguales
H1: los modelos son distintos

Regla:

Si F_est > F_critico → se rechaza H0
Si F_est ≤ F_critico → no se rechaza H0

Conclusión:

Si F_est > F_critico → los modelos difieren entre sectores.
Si F_est ≤ F_critico → no hay evidencia de diferencias entre sectores.
